**Importing libraries**

```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set_style('darkgrid')
```

importing the dataset

```python
df=pd.read_csv('zomato.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop(['menu_item'], 
        axis = 1, inplace=True)
```

```python
df.head()
```

```python
df['rate'].isnull().sum()
```

```python
df['rate'].unique()
```

```python
df['rate'] = df['rate'].fillna(df['rate'].mode()[0])
```

```python
df['rate'].isnull().sum()
```

```python
rate_hyphen = df.loc[df['rate'] == '-'].index
df.drop(rate_hyphen, axis = 0, inplace = True)
df_rate_new = df[df.rate == 'NEW']
df['rate'].replace({"NEW" : "3.0 /5"}, inplace = True)
df['rate'] = df['rate'].apply(lambda x : x.split('/')[0])
df['rate'] = df['rate'].apply(lambda x : x.split(' ')[0])
df = df.astype({'rate' : float})
np.sort(df['rate'].unique())
```

```python
print(df['location'].unique())
print(len(df['location'].unique()))
```

Dropping the rows with null values in location

```python
df = df.dropna(subset= ['location'])
```

Clubbing all the blocks of Koramangala under one location

```python
df_kor = df[df['location'].str.contains('Koramangala')]
df_kor_list = df_kor['location'].unique()
df['location'] = df['location'].replace(df_kor_list, 'Koramangala')
```

```python
df_kor = df[df['listed_in(city)'].str.contains('Koramangala')]
df_kor_list = df_kor['listed_in(city)'].unique()
df['listed_in(city)'] = df['listed_in(city)'].replace(df_kor_list, 'Koramangala')
```

```python
print(df['rest_type'].unique())
```

```python
df = df.dropna(subset= ['rest_type'])
```

```python
df['rest_type'].isnull().sum()
```

```python
print(len(df['dish_liked'].unique()))
```

```python
df['dish_liked'] = df['dish_liked'].fillna('NONE')
```

```python
df['dish_liked'].isnull().sum()
```

```python
df = df.dropna(subset= ['cuisines'])
```

```python
df['cuisines'].isnull().sum()
```

```python
df = df.rename(columns={'approx_cost(for two people)': 'cost'})
df['cost'] = [float(str(i).replace(",", "")) for i in df['cost']]
```

```python
df = df.dropna(subset= ['cost'])
```

```python
df['cost'].isnull().sum()
```

```python
df.head()
df.to_csv('zomato_clean.csv')
```

```python
def Encode(zomato_df):
    for column in zomato_df.columns[~zomato_df.columns.isin(['rate', 'cost', 'votes'])]:
        zomato_df[column] = zomato_df[column].factorize()[0]
    return zomato_df

zomato_en = Encode(df.copy())
zomato_en.head()
```

```python
#Get Correlation between different variables
corr = zomato_en.corr(method='kendall')
plt.figure(figsize=(15,8))
sns.heatmap(corr, annot=True)
zomato_en.columns
```
